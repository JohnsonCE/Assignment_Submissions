Customers with no orders, using customers.csv and orders.csv

```python
from pyspark.sql import SparkSession
from pyspark.sql import functions as F

spark = SparkSession.builder.appName("CustomersWithoutOrders").getOrCreate()

customers = spark.read.csv("../customers.csv", header=True, inferSchema=True)
orders = spark.read.csv("../orders.csv", header=True, inferSchema=True)

customers_without_orders = customers.join(
    orders, on="customer_id", how="left_anti"
)

customers_without_orders.orderBy("customer_id").show(truncate=False)
```
